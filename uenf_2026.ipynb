```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import xarray as xr
import os # Import os module
import cdsapi # Import cdsapi

# Define the filename for the downloaded data
netcdf_file = 'europe_temperature.nc'

# Check if the .nc file exists, if not, download it
if not os.path.exists(netcdf_file):
    print(f"File '{netcdf_file}' not found. Initiating download...")
    dataset = "derived-era5-pressure-levels-daily-statistics"
    request = {
        "product_type": "reanalysis",
        "variable": ["temperature"],
        "year": "2024",
        "month": ["01"],
        "day": ["01"],
        "pressure_level": ["100"],
        "daily_statistic": "daily_mean",
        "time_zone": "utc+12:00",
        "frequency": "6_hourly",
        "area": [70, -15, 35, 30] # North, West, South, East for Europe
    }
    client = cdsapi.Client(url='https://cds.climate.copernicus.eu/api', key='18b27a08-3ae3-4fd1-96ff-f30a52f1404d', verify=False)
    client.retrieve(dataset, request).download(netcdf_file) # Specify filename for download
    print(f"Download of '{netcdf_file}' complete.")
else:
    print(f"File '{netcdf_file}' already exists. Skipping download.")

# Load the dataset from the .nc file
ds = xr.open_dataset(netcdf_file, engine='h5netcdf')

# Select the first time step and pressure level for plotting a 2D map
temp_data_2d = ds['t'].isel(time=0, pressure_level=0)

# Create a figure and an axes object with a Cartopy projection
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

# Set the extent to focus on Europe
ax.set_extent([-15, 30, 35, 70], crs=ccrs.PlateCarree()) # [west, east, south, north]

# Plot the temperature data
c = temp_data_2d.plot.pcolormesh(ax=ax, transform=ccrs.PlateCarree(), cmap='coolwarm', add_colorbar=False)

# Add map features
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)

# Add a colorbar
cbar = plt.colorbar(c, ax=ax, orientation='vertical', pad=0.05, aspect=50)
cbar.set_label('Temperature (K)')

# Set title
plt.title(f'Mean Temperature over Europe at {temp_data_2d.pressure_level.values}hPa on {temp_data_2d.time.dt.strftime('%Y-%m-%d %H:%M').item()}')
plt.show()
```

### Criando um Mapa para a Nova Região (Brasil)

Com as novas coordenadas fornecidas, vamos agora focar em uma nova área geográfica. Primeiro, vamos definir os parâmetros para baixar os dados climáticos para esta nova região usando a API CDS. Em seguida, carregaremos esses dados e geraremos um mapa de calor semelhante ao da Europa.

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import xarray as xr
import os # Import os module
import cdsapi # Import cdsapi

# Definir as coordenadas para a nova região (Brasil)
# Latitude Norte (máx): -1.75
# Latitude Sul (mín): -2.25
# Longitude Oeste (mín): -49.20
# Longitude Leste (máx): -48.70
new_area_coords = [-1.75, -49.20, -2.25, -48.70] # [North, West, South, East]

# Definir o nome do arquivo para os dados baixados da nova região
new_netcdf_file = 'brazil_temperature.nc'

# Check if the .nc file exists, if not, download it
if not os.path.exists(new_netcdf_file):
    print(f"File '{new_netcdf_file}' not found. Initiating download...")
    dataset = "derived-era5-pressure-levels-daily-statistics"
    request = {
        "product_type": "reanalysis",
        "variable": ["temperature"],
        "year": "2024",
        "month": ["01"],
        "day": ["01"],
        "pressure_level": ["100"],
        "daily_statistic": "daily_mean",
        "time_zone": "utc+12:00",
        "frequency": "6_hourly",
        "area": new_area_coords
    }
    client = cdsapi.Client(url='https://cds.climate.copernicus.eu/api', key='18b27a08-3ae3-4fd1-96ff-f30a52f1404d', verify=False)
    client.retrieve(dataset, request).download(new_netcdf_file) # Specify filename for download
    print(f"Download of '{new_netcdf_file}' complete.")
else:
    print(f"File '{new_netcdf_file}' already exists. Skipping download.")

# Carregar o dataset do arquivo .nc da nova região
ds_brazil = xr.open_dataset(new_netcdf_file, engine='h5netcdf')

# Selecionar o primeiro passo de tempo e nível de pressão para plotar um mapa 2D
temp_data_brazil_2d = ds_brazil['t'].isel(valid_time=0, pressure_level=0)

# Criar uma figura e um objeto axes com uma projeção Cartopy
fig_brazil = plt.figure(figsize=(15, 12)) # Aumentando o tamanho da figura para melhor resolução visual
ax_brazil = fig_brazil.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

# Definir a extensão para focar na nova região (Brasil)
# [oeste, leste, sul, norte] - note que 'west' e 'south' são negativos
ax_brazil.set_extent([new_area_coords[1], new_area_coords[3], new_area_coords[2], new_area_coords[0]], crs=ccrs.PlateCarree())

# Plotar os dados de temperatura como um mapa de calor
heatmap_brazil = temp_data_brazil_2d.plot.pcolormesh(
    ax=ax_brazil,
    transform=ccrs.PlateCarree(),
    cmap='viridis', # Escolha um mapa de cores apropriado para temperatura
    add_colorbar=False # Adicionaremos uma colorbar personalizada
)

# Adicionar funcionalidades ao mapa
ax_brazil.add_feature(cfeature.COASTLINE, linewidth=0.8)
ax_brazil.add_feature(cfeature.BORDERS, linestyle=':', edgecolor='gray', linewidth=0.6)
ax_brazil.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False, color='lightgray', alpha=0.5)

# Adicionar uma colorbar
cbar_brazil = plt.colorbar(heatmap_brazil, ax=ax_brazil, orientation='vertical', pad=0.05, aspect=50)
cbar_brazil.set_label('Temperatura (K)', fontsize=12)

# Definir título
plt.title(
    f'Mapa de Calor da Temperatura Média sobre o Brasil (Região Especificada) a {temp_data_brazil_2d.pressure_level.values}hPa em {temp_data_brazil_2d.valid_time.dt.strftime('%Y-%m-%d %H:%M').item()}',
    fontsize=14
)
plt.show()
```

### Download de Dados para a Região do Brasil com Parâmetros Atualizados

Agora, vamos baixar os dados para a mesma região do Brasil, mas com um conjunto diferente de meses e nível de pressão, conforme solicitado.


```python
import cdsapi
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import os

# Definir as coordenadas para a região (Brasil)
# [North, West, South, East]
updated_area_coords = [-1.75, -49.20, -2.25, -48.70]

# Definir o nome do arquivo para os dados baixados com os novos parâmetros
updated_netcdf_file = 'brazil_temperature_updated_params.nc'

# Configurar a solicitação com os novos parâmetros
# Meses: 01, 02, 03, 09, 10, 11, 12
# Nível de Pressão: 1000 hPa

dataset_updated = "derived-era5-pressure-levels-daily-statistics"
request_updated = {
    "product_type": "reanalysis",
    "variable": ["temperature"],
    "year": "2024",
    "month": ["01", "02", "03", "09", "10", "11", "12"],
    "day": ["01"], # Mantendo o dia 01 para simplificar, mas pode ser ajustado conforme necessário
    "pressure_level": ["1000"], # Nível de pressão atualizado
    "daily_statistic": "daily_mean",
    "time_zone": "utc+12:00",
    "frequency": "6_hourly",
    "area": updated_area_coords
}

# Verificar se o arquivo .nc já existe, se não, baixá-lo
if not os.path.exists(updated_netcdf_file):
    print(f"File '{updated_netcdf_file}' not found. Initiating download with updated parameters...")
    client_updated = cdsapi.Client(url='https://cds.climate.copernicus.eu/api', key='18b27a08-3ae3-4fd1-96ff-f30a52f1404d', verify=False)
    client_updated.retrieve(dataset_updated, request_updated).download(updated_netcdf_file)
    print(f"Download of '{updated_netcdf_file}' complete.")
else:
    print(f"File '{updated_netcdf_file}' already exists. Skipping download.")

# Carregar o dataset do arquivo .nc da nova região com parâmetros atualizados
ds_brazil_updated = xr.open_dataset(updated_netcdf_file, engine='h5netcdf')

# Selecionar o primeiro passo de tempo e nível de pressão para plotar um mapa 2D
# Nota: Como agora temos múltiplos meses, vamos pegar o primeiro 'valid_time' disponível.
temp_data_brazil_updated_2d = ds_brazil_updated['t'].isel(valid_time=0, pressure_level=0)

# Criar uma figura e um objeto axes com uma projeção Cartopy
fig_brazil_updated = plt.figure(figsize=(15, 12)) # Mantendo o tamanho da figura para melhor resolução visual
ax_brazil_updated = fig_brazil_updated.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

# Definir a extensão para focar na região do Brasil
ax_brazil_updated.set_extent([updated_area_coords[1], updated_area_coords[3], updated_area_coords[2], updated_area_coords[0]], crs=ccrs.PlateCarree())

# Plotar os dados de temperatura como um mapa de calor
heatmap_brazil_updated = temp_data_brazil_updated_2d.plot.pcolormesh(
    ax=ax_brazil_updated,
    transform=ccrs.PlateCarree(),
    cmap='viridis', # Escolha um mapa de cores apropriado para temperatura
    add_colorbar=False
)

# Adicionar funcionalidades ao mapa
ax_brazil_updated.add_feature(cfeature.COASTLINE, linewidth=0.8)
ax_brazil_updated.add_feature(cfeature.BORDERS, linestyle=':', edgecolor='gray', linewidth=0.6)
ax_brazil_updated.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False, color='lightgray', alpha=0.5)

# Adicionar uma colorbar
cbar_brazil_updated = plt.colorbar(heatmap_brazil_updated, ax=ax_brazil_updated, orientation='vertical', pad=0.05, aspect=50)
cbar_brazil_updated.set_label('Temperatura (K)', fontsize=12)

# Definir título
plt.title(
    f'Mapa de Calor da Temperatura Média sobre o Brasil (Parâmetros Atualizados) a {temp_data_brazil_updated_2d.pressure_level.values}hPa em {temp_data_brazil_updated_2d.valid_time.dt.strftime('%Y-%m-%d %H:%M').item()}',
    fontsize=14
)
plt.show()
```

### Média de Temperatura Mensal para a Região do Brasil (Parâmetros Atualizados)

Agora, vamos calcular a temperatura média para cada mês disponível no conjunto de dados `ds_brazil_updated`.

```python
# Calcule a média de temperatura para cada mês
# Primeiro, garantimos que a variável 't' esteja disponível no dataset
if 't' in ds_brazil_updated.data_vars:
    # Resample os dados por mês ('MS' para início do mês) e calcule a média
    monthly_mean_temperature = ds_brazil_updated['t'].resample(valid_time='MS').mean(dim='valid_time')

    print("Média de Temperatura Mensal (K) para a região do Brasil:")
    display(monthly_mean_temperature)

    # Converter para Celsius
    monthly_mean_temperature_celsius = monthly_mean_temperature - 273.15
    print("\nMédia de Temperatura Mensal (°C):")
    display(monthly_mean_temperature_celsius)
else:
    print("A variável 't' (temperatura) não foi encontrada no dataset ds_brazil_updated.")
```

### Gráfico de Variação da Temperatura Média Mensal

Vamos agora visualizar a variação da temperatura média mensal em Celsius por meio de um gráfico de linha.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Certificar-me de que monthly_mean_temperature_celsius está disponível e tem a dimensão 'valid_time'
if 'monthly_mean_temperature_celsius' in locals() and 'valid_time' in monthly_mean_temperature_celsius.dims:
    # Para simplificar a plotagem, podemos pegar a média espacial também para ter uma única linha por mês
    # Ou podemos plotar cada ponto de latitude/longitude individualmente, mas para uma 'variação mensal', uma média geral é mais comum.
    # Aqui, vou calcular a média sobre latitude e longitude para obter uma única temperatura média por mês.
    monthly_mean_temp_spatial_avg = monthly_mean_temperature_celsius.mean(dim=['latitude', 'longitude', 'pressure_level'])

    # Converter o xarray DataArray para um pandas Series para facilitar a plotagem se necessário
    monthly_mean_temp_series = monthly_mean_temp_spatial_avg.to_series()

    plt.figure(figsize=(12, 6))
    monthly_mean_temp_series.plot(kind='line', marker='o')
    plt.title('Variação da Temperatura Média Mensal (2024)')
    plt.xlabel('Mês')
    plt.ylabel('Temperatura Média (°C)')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.xticks(rotation=45) # Rotaciona os rótulos do eixo x para melhor leitura
    plt.tight_layout() # Ajusta o layout para evitar sobreposição
    plt.show()
else:
    print("Não foi possível gerar o gráfico. 'monthly_mean_temperature_celsius' não está disponível ou não possui a dimensão 'valid_time'.")
```

### Mapa de Localização de Igarapé-Miri (PA)

Agora, vamos gerar um mapa para visualizar a localização de Igarapé-Miri no Pará, destacando um ponto específico e a área do município com um basemap do OpenStreetMap.

```python
# Instale as bibliotecas (caso necessário)
!pip install cartopy contextily geopandas matplotlib mercantile xyzservices matplotlib-scalebar

import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point, box
import contextily as ctx
from matplotlib_scalebar.scalebar import ScaleBar

# -----------------------------
# Coordenadas de Igarapé-Miri
# -----------------------------
lon = -48.96
lat = -1.97

# Cria ponto
gdf = gpd.GeoDataFrame(
    geometry=[Point(lon, lat)],
    crs="EPSG:4326"
)

# Converte para Web Mercator
gdf = gdf.to_crs(epsg=3857)

x = gdf.geometry.x.iloc[0]
y = gdf.geometry.y.iloc[0]

# -----------------------------
# QUADRADO (área do município)
# -----------------------------
lon_min = -49.20
lon_max = -48.70
lat_min = -2.25
lat_max = -1.75

quad = gpd.GeoDataFrame(
    geometry=[box(lon_min, lat_min, lon_max, lat_max)],
    crs="EPSG:4326"
).to_crs(epsg=3857)

# -----------------------------
# MAPA
# -----------------------------
fig, ax = plt.subplots(figsize=(10,10))

# Ponto
ax.scatter(
    x, y,
    color='red',
    s=70,
    edgecolor='black',
    zorder=5
)

ax.text(
    x + 2000,
    y + 2000,
    "Igarapé-Miri",
    fontsize=11,
    color='red',
    weight='bold'
)

# Quadrado do município
quad.boundary.plot(
    ax=ax,
    edgecolor="red",
    linewidth=2.5,
    zorder=4
)

# Extensão do mapa
dist = 60000
ax.set_xlim(x - dist, x + dist)
ax.set_ylim(y - dist, y + dist)

# Fundo OSM
ctx.add_basemap(
    ax,
    source=ctx.providers.OpenStreetMap.Mapnik
)

# -----------------------------
# ESCALA DO MAPA
# -----------------------------
scalebar = ScaleBar(
    1,               # 1 pixel = 1 unidade do CRS (Web Mercator)
    units="m",
    dimension="si-length",
    location="lower left",
    box_alpha=0.6,
    color="black"
)

ax.add_artist(scalebar)

# -----------------------------
# Finalização
# -----------------------------
ax.set_axis_off()

plt.title(
    "Mapa de localização de Igarapé-Miri (PA)",
    fontsize=14,
    weight='bold'
)

plt.tight_layout()
plt.savefig(
    "Mapa_Localizacao_Igarape_Miri_OSM.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, box
import contextily as ctx

# -----------------------------
# 1. CARREGAR DADOS
# -----------------------------
df = pd.read_csv("file.csv")

# temperatura em Celsius
df["temp_c"] = df["t"] - 273.15

# garantir tipos corretos
df["latitude"] = df["latitude"].astype(float)
df["longitude"] = df["longitude"].astype(float)

# -----------------------------
# 2. MÉDIA ESPACIAL
# -----------------------------
df_mean = df.groupby(["latitude", "longitude"], as_index=False)["temp_c"].mean()

# -----------------------------
# 3. GEO DATAFRAME (WGS84)
# -----------------------------
gdf = gpd.GeoDataFrame(
    df_mean,
    geometry=[Point(xy) for xy in zip(df_mean["longitude"], df_mean["latitude"])],
    crs="EPSG:4326"
)

# -----------------------------
# 4. ÁREA DO MUNICÍPIO
# -----------------------------
lon_min, lon_max = -49.20, -48.70
lat_min, lat_max = -2.25, -1.75

quad = gpd.GeoDataFrame(
    geometry=[box(lon_min, lat_min, lon_max, lat_max)],
    crs="EPSG:4326"
)

# -----------------------------
# 5. PROJETAR PARA WEB MERCATOR (OBRIGATÓRIO)
# -----------------------------
gdf = gdf.to_crs(epsg=3857)
quad = quad.to_crs(epsg=3857)

# limites do mapa em 3857
minx, miny, maxx, maxy = quad.total_bounds

# -----------------------------
# 6. CRIAR MAPA
# -----------------------------
fig, ax = plt.subplots(figsize=(10,10))

ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)

# -----------------------------
# 7. BASE MAPA (OSM)
# -----------------------------
ctx.add_basemap(
    ax,
    source=ctx.providers.CartoDB.Positron
)

# -----------------------------
# 8. PONTOS DE TEMPERATURA
# -----------------------------
gdf.plot(
    ax=ax,
    column="temp_c",
    cmap="coolwarm",
    markersize=250,
    legend=True,
    edgecolor="black",
    alpha=0.85
)

# -----------------------------
# 9. CONTORNO DO MUNICÍPIO
# -----------------------------
quad.boundary.plot(
    ax=ax,
    color="black",
    linewidth=2
)

# -----------------------------
# 10. FINALIZAÇÃO
# -----------------------------
ax.set_title("Temperatura média ERA5 - Igarapé-Miri (2024)", fontsize=14)
ax.set_axis_off()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from scipy.interpolate import griddata
```

```python
df = pd.read_csv("file.csv")

df["temp_c"] = df["t"] - 273.15

df["latitude"] = df["latitude"].astype(float)
df["longitude"] = df["longitude"].astype(float)
```

```python
df_mean = df.groupby(["latitude", "longitude"], as_index=False)["temp_c"].mean()
```

```python
x = df_mean["longitude"].values
y = df_mean["latitude"].values
z = df_mean["temp_c"].values
```

```python
xi = np.linspace(-49.2, -48.7, 200)
yi = np.linspace(-2.25, -1.75, 200)

xi, yi = np.meshgrid(xi, yi)
```

```python
zi = griddata(
    (x, y),
    z,
    (xi, yi),
    method="linear"
)

# evitar buracos
zi = np.nan_to_num(zi, nan=np.nanmean(z))
```

```python
fig = plt.figure(figsize=(10,10))

ax = plt.axes(projection=ccrs.PlateCarree())

ax.set_extent([-49.2, -48.7, -2.25, -1.75], crs=ccrs.PlateCarree())
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection=ccrs.PlateCarree())
```

```python
ax.set_extent([-49.2, -48.7, -2.25, -1.75], crs=ccrs.PlateCarree())
```

```python
ax.coastlines(resolution="10m", linewidth=0.5)
ax.add_feature(cfeature.BORDERS, linewidth=0.5)
```

```python
mesh = ax.contourf(
    xi, yi, zi,
    levels=12,
    cmap="coolwarm",
    transform=ccrs.PlateCarree()
)
```

```python
cbar = plt.colorbar(mesh, ax=ax, shrink=0.7)
cbar.set_label("Temperatura média (°C)")
```

```python
ax.scatter(
    -48.96, -1.97,
    color="black",
    s=60,
    transform=ccrs.PlateCarree(),
    zorder=5
)

ax.text(
    -48.96, -1.97,
    "Igarapé-Miri",
    transform=ccrs.PlateCarree(),
    fontsize=10,
    weight="bold"
)
```

```python
ax.set_title("Mapa de Contorno ERA5 - Igarapé-Miri (2024)")
plt.tight_layout()
plt.show()
```

```python
temp = (
    df.groupby(["latitude", "longitude"])["temp_c"]
      .mean()
      .unstack()
)

print(temp)
```

```python
lon = temp.columns.values
lat = temp.index.values
temp = temp.values
```

```python
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection=ccrs.PlateCarree())

ax.set_extent(
    [-49.2, -48.7,
     -2.3, -1.7],
    crs=ccrs.PlateCarree()
)

cf = ax.contourf(
    lon,
    lat,
    temp,
    levels=10,
    cmap="coolwarm",
    transform=ccrs.PlateCarree()
)

cs = ax.contour(
    lon,
    lat,
    temp,
    colors="black",
    linewidths=0.7,
    transform=ccrs.PlateCarree()
)

ax.clabel(cs, fontsize=8)

ax.scatter(
    -48.96,
    -1.97,
    color="red",
    s=80,
    transform=ccrs.PlateCarree()
)

ax.text(
    -48.96,
    -1.97,
    " Igarapé-Miri",
    transform=ccrs.PlateCarree()
)

plt.colorbar(cf, ax=ax, label="Temperatura média (°C)")

ax.set_title(
    "Temperatura Média ERA5 (2024)\nIgarapé-Miri - PA"
)

plt.show()
```
